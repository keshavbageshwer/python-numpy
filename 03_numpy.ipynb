# Advanced operations with real example

````python
import numpy as np 
import matplotlib.pyplot as plt
````

````python
# data struture : [restro_id,2021,2022,2023,2024]
sales_data=np.array([
    [1,1500,3000,5000,20000],
    [2,122,1000,2000,20300],
    [3,245,535,765,876],
    [4,2556,55354,75665,45876],
    [5,299,5352,76455,87256],

])

print('==sales analysis==')
print('\n sales data shape',sales_data.shape)
print('\nfirst three rows of data\n',sales_data[0:3])
print('\n to print all data but not id\n',sales_data[:,1:])

# [:,1:] the 2d matrix takes like [rows, column] [:] means all the rows and [:,1:]means all rows and columns starts from 1 till end 
print('\n to print all data but not id\n',sales_data[:,1:3])
````

````python
# adding a complete row or column 
````

* sum method in numpy

````python
print(np.sum(sales_data,axis=0))
# or 
total_sum=np.sum(sales_data[:,1:],axis=0)
print(total_sum)
````

* min, max & average (mean) method in numpy

````python
min_sale=np.min(sales_data[:,1:], axis=1) # axis 1 means left to right 
print('minimum sales per company:',min_sale)

max_sale=np.max(sales_data[:,1:], axis=0)# axis=0 means top to bottom 
print('max sale per year',max_sale)

# average sales per company 
avg_sale=np.mean(sales_data[:,1:],axis=1)
print('avereage sale per company',avg_sale)
````

* to calculate cummulative sum (cumsum) when the new value is calculated by adding the previous value sum

````python
cumsum=np.cumsum(sales_data[:,1:],axis=1)# means left to right total sales of the company from 2021 to last 
print(cumsum)
````

## Using Matplotlib

````python
plt.figure(figsize=(8,6)) # this 8,6 is a tuple and it denotes the size of the figure/ chart/ graph that will be shown 
plt.plot(np.mean(cumsum,axis=0))# give what you want to plot 
plt.title("average cummulative sales across all companies ")
plt.xlabel('years') # what is written on x axis
plt.ylabel('sales')# what written on y axis
plt.grid(True)
plt.show()# only used with jupyter notebook
````

## Vector Operations

````python
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])
````

1) vector addition

````python
print('vector addition:',vector1+vector2)
````

2) vector multiplication

````python
print('vector multiplication',vector1*vector2)
````

3) vector dot product 
* dot product = (1x6)+(2x7)+(3x8)+(4x9)+(5x10)

````python
print('vector dot product is:',np.dot(vector1,vector2))
````

4) angle between vectors ( you use arccos method)

````python
angle=np.arccos(np.dot(vector1,vector2)/ (np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
````

## np.vctorize

## Understanding `np.vectorize()`

The `np.vectorize()` function takes a standard Python function and transforms it into a vectorized function that can be applied element-wise across a NumPy array.

### Example Usage

```python
import numpy as np

# Define a standard Python function
def square(x):
    return x * x

# Convert it into a vectorized function
vectorized_square = np.vectorize(square)

# Apply it element-wise to an array
result = vectorized_square(np.array([1, 2, 3]))
print(result)
```

**Output:**
```text
[1 4 9]
```
