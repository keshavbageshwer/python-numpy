## Matrix Broadcasting and Difference Matrices

Suppose you have two 1-D arrays, \(A\) and \(B\):

```python
import numpy as np

A = np.array([3, 11, 4, 5])
B = np.array([5, 0, 3])
```

Your goal is to build a difference matrix where each element at position \((i, j)\) gives:
\[C_{i, j} = A_i - B_j\]

In other words, you want to subtract each element of \(B\) from each element of \(A\).

---

### The Challenge with 1-D Arrays

If you try to run `A - B`, NumPy will throw a `ValueError` because the arrays do not have compatible shapes for direct operation:

```text
ValueError: operands could not be broadcast together with shapes (4,) (3,)
```

Even if they were the exact same size, NumPy would only perform simple element-wise subtraction instead of generating a full cross-difference matrix.

---

### The Broadcasting Solution

If \(A\) is transformed into a \(4 x 1\) array and \(B\) is transformed into a \(1 x 3\) array, NumPy broadcasting rules will automatically expand both arrays to produce the desired difference matrix.

#### 1. Reshape Array \(A\)
Convert \(A\) from shape `(4,)` into a `(4, 1)` column vector using `np.newaxis`:

```python
A[:, np.newaxis]
```
**Output:**
```python
array([[ 3],
,
       [ 4],
       [ 5]])
```

#### 2. Reshape Array \(B\)
Convert \(B\) from shape `(3,)` into a `(1, 3)` row vector:

```python
B[np.newaxis, :]
```
**Output:**
```python
array([[5, 0, 3]])
```

#### 3. Calculate the Difference Matrix

```python
A[:, np.newaxis] - B[np.newaxis, :]
```
**Output:**
```python
array([[-2,  3,  0],
       [ 6, 11,  8],
       [-1,  4,  1],
       [ 0,  5,  2]])
```

---

### Shortcuts & Simplifications

#### Vector-to-Array Broadcasting
You can simplify the syntax to `A[:, np.newaxis] - B`. NumPy broadcasting automatically treats the 1-D array \(B\) as a row vector to match the 2-D column vector:

```python
A[:, np.newaxis] - B
```

#### Using `None`
`np.newaxis` is an alias for `None`. You can write the expression even more concisely like this:

```python
A[:, None] - B
```

# Boolean Indexing

With boolean indexing, you can subset an array `A` using another array `B` composed of boolean values (`True`/`False`).

---

## Example Walkthrough

### 1. Define the Initial Array
Suppose we have a \(3 \times 3\) array named `foo`:

```python
import numpy as np

foo = np.array([,
 ,
    [3, 3, 1]
])
```

---

### 2. Create a Comparison Mask
We can generate a boolean mask to find everywhere the value equals `3`:

```python
mask = foo == 3
print(mask)
```

**Output:**
```python
[[ True False False]
 [False False  True]
 [ True  True False]]
```

---

### 3. Subset the Array Using the Mask
We can apply this `mask` back to `foo` to isolate and extract only the elements that match our condition:

```python
print(foo[mask])
```

**Output:**
```python
[3 3 3 3]
```

---

### 4. Modify Values In-Place
Furthermore, we can use the exact same `mask` to modify elements conditionally—for example, converting all `3`s in `foo` to `0`s:

```python
foo[mask] = 0
print(foo)
```

**Output:**
```python
[[0 9 7]
 [2 0 0]
 [0 0 1]]
```


````python
foo = np.array([
    [3, 9, 7],
    [2, 0, 3],
    [3, 3, 1]
])
````

````python
r13 = np.array([True, False, True])
c23 = np.array([False, True, True])
````

````python
# to get the columns 
print(foo[:,c23])


# to get the rows 
print(foo[r13])
````
