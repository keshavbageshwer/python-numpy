## Array Reshaping 

````python
arr = np.arange(12)
print("Original array ", arr) # this will generate an array from 0 to 11 

reshaped = arr.reshape((3, 4)) # this will reshape the array 
print("\n Reshaped array ", reshaped)

# note:- the array can only be reshaped according to the numbr of elements it have as mxn array can be converted only by the array having m*n elements 
# example a array having 12 elements can only generate an array with the dimension of 3x4 or 4x3 or 6x2 or 2x6 or 1x12 or 12x1 

flattened = reshaped.flatten()
print("\n Flattened array ", flattened)
# this basically converts the array into 1-d array that means flattening the array 

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)


#[[1,2,3,4,5,6]] this is treated like a 1x6 matrix in python
# and [1,2,3,4,5,6] this is an 1-d array  

# Transpose
transpose = reshaped.T # it does the transpose of the reshaped matrix 
print("\n Transposed array ", transpose) # transpos of a matrix is nothing but making a mxn matrix into nxm 
````

# NumPy: `flatten()` vs `ravel()`

The primary difference between `flatten()` and `ravel()` is memory allocation:
* **`flatten()`** always returns a **copy** of the original array.
* **`ravel()`** returns a **view** (reference) of the original array whenever possible.

---

## 📊 Quick Comparison

| Feature | `array.flatten()` | `array.ravel()` / `np.ravel(array)` |
| :--- | :--- | :--- |
| **Memory** | Always allocates new memory (**Copy**) | Shares memory with the original array (**View**) |
| **Data Mutation** | Modifying the output **does not** affect the original | Modifying the output **will change** the original |
| **Performance** | Slower (requires copying data blocks) | Faster (avoids data duplication) |
| **Syntax Type** | Only available as an `ndarray` method | Available as a function and an `ndarray` method |
| **Input Flexibility**| Requires a strict NumPy array | Accepts non-array structures (e.g., standard lists) |


## 💻 Code Examples

### 1. Modifying a `flatten()` Array (Safe & Isolated)
Because `flatten()` generates a distinct deep copy, changing its elements leaves the parent matrix completely untouched.
```python
import numpy as np

original = np.array([[1, 2], [3, 4]])
flat_arr = original.flatten()

# Modify the 1D array
flat_arr[0] = 99

print("Flattened:", flat_arr)
print("Original:\n", original)  # Unchanged
```

### 2. Modifying a `ravel()` Array (Shared Memory)
Because `ravel()` defaults to a view, mutating the 1D output directly updates the source matrix.

```python
import numpy as np

original = np.array([[1, 2], [3, 4]])
ravel_arr = original.ravel()

# Modify the 1D array
ravel_arr[0] = 99

print("Raveled:", ravel_arr)
print("Original:\n", original)  # Original changes!
```
*> **Note:** If the original array consists of non-contiguous memory slices, `ravel()` will automatically fallback to creating a copy instead of a view.*

---

## 💡 Summary Rule of Thumb
* Use **`ravel()`** for read-only loops, math metrics, or when you explicitly want changes to sync back to the main matrix. It optimizes speed and memory.
* Use **`flatten()`** when you plan to manipulate the 1D array separately and must guarantee that your original multi-dimensional dataset remains protected.


````python

#in detail ravel 


# Create a 2D array (2 rows, 3 columns)
matrix = np.array([[1, 2, 3], 
                   [4, 5, 6]])

# Flatten the array
flat_view = np.ravel(matrix)
print(flat_view)  
# Output: [1 2 3 4 5 6]

# Modifying the view changes the original matrix!
flat_view[0] = 99
print(matrix)
# Output: 
# [[99  2  3]
#  [ 4  5  6]]
````
